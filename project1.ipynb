```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import scipy.stats as stats
```

```python
df=pd.read_csv('weatherHistory.csv')
df.head(15)
```

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
```

```python
df.info()
```

```python
df['Formatted Date']
```

```python
df.head()
```

```python
df=df.set_index('Formatted Date') #index
df.head()
```

```python
data_sampling=df[['Apparent Temperature (C)','Humidity']].resample('M').mean()
```

```python
data_sampling.head(20)
```

```python
plt.figure(figsize=(15,4))
plt.title("Relation between Temperature and Humidity",fontsize='17')
plt.xlabel("Graph according to each year",fontsize='14')
plt.plot(data_sampling['Apparent Temperature (C)'],label='Apparent Temperature (C)',color='red')
plt.plot(data_sampling['Humidity'],label='Humidity',color='blue')
plt.legend(fontsize='10')
plt.show()
```

From above graph we can see that Humidity remains almost constant in each year.

```python
test=stats.ttest_ind(data_sampling['Apparent Temperature (C)'],data_sampling['Humidity'])
result=test[1]
print(result)
```

```python
if result>0.05:
    print ("Accept  null hypothesis")
elif result<0.05:
    print ("Reject null hypothesis")
```
